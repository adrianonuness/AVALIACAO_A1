```python
'''
Sistema simples de cadastro de produtos com nome e preço, armazenando em
dicionário.
'''
#comecei adicionando um dicionário (vazio) de produtos representados por chaves.
produtos = {}
#Loop de infinito para informar o nome do produto
while True:
  nome = input("Informe o nome do produto ou Enter para sair: ")
  if nome == "":
    break #Se o usuário dar Enter sem ter escrito nada, o loop é encerrado.

  #Aqui o usuário informa o preço do produto
  preco = float(input("Informe o preço do produto: "))


  produtos[nome] = preco #Aqui é onde o produto com nome e preço é armazenado
                         #no dicionário(produtos).

  print("Produto cadastrado.")

print("Lista dos produtos cadastrados: ")
print(produtos)
```
