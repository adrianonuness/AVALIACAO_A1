```python
'''
Programa que receba nomes de alunos e armazene em uma lista,
permitindo exibir todos os nomes ao final.
'''
alunos = []

#Coloco um loop infinito usando while True, para que o laço seja desfeito
#apenas quando eu usar o break, em alguma condição específica:
while True:
 nome = input("Digite o nome do aluno(a) ou então aperte Enter para parar: ")

 if nome == "" :
  break
#aqui eu coloquei a condição if, se o usuário apertar Enter sem ter digitado
#nada (""), o break será ativado, encerrando o loop.

 alunos.append(nome)
#Aqui os nomes da variável(nome) digitados serão adicionados à lista(alunos),
#através do .append

#agora basta printar a lista(alunos)
print("Lista dos alunos informados:")
print(alunos)
```
