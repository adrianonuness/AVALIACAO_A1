```python
'''
PEDRA, PAPEL, TESOURA(JOKENPÔ)
Fiz um código simples de python (Com as funções que eu sei usar no momento,
usando lista, biblioteca, loop while e condições if, elif e else.)
'''
#comecei importando a biblioteca random para que o computador escolha uma opção
#na lista dada de forma aleatória.
import random
lista = ["pedra", "papel", "tesoura"]

#loop para que o código continue em execução, sem que o usuário tenha que clicar
#em executar para continuar jogando
while True:
 usuario = input("escolha entre as opções (PEDRA, PAPEL OU TESOURA) ou digite sair para parar: ")
 computador = random.choice(lista)
 #usuário escolhe as opções do jogo ou digita sair para parar de jogar
 #computador escolhe uma opção aleatória dentro da lista dada

 #Daqui em diante eu coloquei um .lower() na variável usuario, isso serve
 #para que o texto leia tudo digitado como letra minúscula, evitando erros
 if usuario.lower() not in lista and usuario.lower() != "sair":
  print("Opção inválida")
#se o usuário digitar algo que não está nas opções, aparecerá uma mensagem de ("opção inválida")

#Aqui abaixo são as condicionais, usei if e elif para os casos em que o usuário vença e
#caso haja empate. Usei else para casos em que o computador vença.
 elif usuario.lower() == "sair":
  print("Jogo encerrado. ")
  break #Aqui o jogo é encerrado se o usuário digitar "sair"

 elif usuario.lower() == computador.lower():
  print(f"O computador escolheu {computador}. EMPATE!! ")
  #coloquei f" para a variavel computador ser revelada caso seja
  #igual a variavel usuario, dando um EMPATE

 elif usuario.lower() == "pedra" and computador.lower() == "tesoura":
  print("O computador escolheu Tesoura. Você venceu! ")
 elif usuario.lower() == "papel" and computador.lower() == "pedra":
  print("O computador escolheu Pedra. Você venceu! ")
 elif usuario.lower() == "tesoura" and computador.lower() == "papel":
  print("O computador escolheu Papel. Você venceu! ")
 else:
  print(f" O computador escolheu {computador}. VOCÊ PERDEU !!!")
  #O else vai ser todas as possibilidades de derrota do usuário, exibindo a
  #Mensagem "VOCÊ PERDEU !!!"
```
